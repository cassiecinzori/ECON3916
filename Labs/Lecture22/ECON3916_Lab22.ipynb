````python
# ------------------------------
# GUIDED - Run as-is
# Install required packages (Colab-safe)
# ------------------------------
!pip install wbgapi scikit-learn matplotlib seaborn -q

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

import wbgapi as wb
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score, silhouette_samples

np.random.seed(42)

print('Libraries loaded. Ready to cluster economies.')
````

## Part 1: Load World Bank WDI Data (GUIDED)

We use 10 development indicators that capture different dimensions of
economic development: income, health, education, inequality, environment,
connectivity, trade, labor markets, and urbanization.

The `wbgapi` package provides direct access to the World Bank's data API.

````python
# ------------------------------
# GUIDED - Run as-is
# Step 1: Download 10 development indicators for all countries
# ------------------------------

indicators = {
    'NY.GDP.PCAP.PP.CD': 'gdp_per_capita_ppp',
    'SP.DYN.LE00.IN': 'life_expectancy',
    'SP.DYN.IMRT.IN': 'infant_mortality',
    'SE.PRM.ENRR': 'primary_enrollment',
    'SI.POV.GINI': 'gini_index',
    'EN.ATM.CO2E.PC': 'co2_per_capita',
    'IT.NET.USER.ZS': 'internet_users_pct',
    'NE.TRD.GNFS.ZS': 'trade_pct_gdp',
    'SL.UEM.TOTL.ZS': 'unemployment_rate',
    'SP.URB.TOTL.IN.ZS': 'urban_population_pct'
}

feature_names = list(indicators.values())

# Fetch most recent available values
df = wb.data.DataFrame(list(indicators.keys()), mrv=1, labels=True)
df = df.rename(columns=indicators)

# Drop countries with too much missing data, then impute remainder
df = df.dropna(thresh=7)  # Keep countries with at least 7 of 10 indicators
df = df.fillna(df.median())  # Impute remaining NAs with column median

print(f'Countries retained: {len(df)}')
print(f'Features ({len(feature_names)}): {feature_names}')
print()
print(df[feature_names].describe().round(1))
````

## Part 2: Standardize Features (GUIDED)

K-Means uses Euclidean distance. Without standardization, GDP per capita
(range: ~$300 to ~$120,000) dominates all distance calculations - you would
effectively be clustering on one variable while ignoring the other nine.

`StandardScaler` transforms each feature to have mean 0 and standard deviation 1:

$$z_{ij} = \frac{x_{ij} - \bar{x}_j}{s_j}$$

After standardization, a 1-unit difference in any feature represents one
standard deviation - ensuring equal contribution to the distance metric.

````python
# ------------------------------
# GUIDED - Run as-is
# Step 2: Standardize features with StandardScaler
# ------------------------------

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[feature_names])

# Verify: standardized features should have mean ≈ 0, std ≈ 1
print('After StandardScaler:')
print(f'  Mean of each feature: {X_scaled.mean(axis=0).round(4)}')
print(f'  Std of each feature:  {X_scaled.std(axis=0).round(4)}')
print()
print(f'Shape of feature matrix: {X_scaled.shape}')
print(f'  {X_scaled.shape[0]} countries x {X_scaled.shape[1]} standardized features')
````

## Part 3: Fit K-Means (K=4) and Visualize with PCA (YOUR TASK)

Now it is your turn. Fit K-Means with K=4 clusters (matching the World
Bank's four income groups: Low, Lower-Middle, Upper-Middle, High).
Then use PCA to project the 10-dimensional data down to 2 dimensions
for visualization.

**Hints:**
- `KMeans(n_clusters=4, init='k-means++', n_init='auto', random_state=42)`
- `PCA(n_components=2)` reduces to 2D - fit on the scaled data
- Color the scatter plot by cluster labels

````python
# ------------------------------
# ✏️ YOUR TASK - Fill in the blanks
# Step 3: Fit K-Means and visualize with PCA
# ------------------------------

# 3a. Fit K-Means with K=4
kmeans = KMeans(
    n_clusters=4,        # Number of clusters (match World Bank income groups)
    init='k-means++',   # Initialization method: k-means++ for smarter centroid seeding
    n_init='auto',
    random_state=42      # For reproducibility
)
df['cluster'] = kmeans.fit_predict(X_scaled)

# Print cluster sizes
print('=== Cluster Sizes ===')
print(df['cluster'].value_counts().sort_index())
print()

# Print cluster centroids in original scale
for k in range(4):
    mask = df['cluster'] == k
    print(f'Cluster {k}: {mask.sum()} countries')
    print(f'  Avg GDP/capita (PPP): ${df.loc[mask, "gdp_per_capita_ppp"].mean():,.0f}')
    print(f'  Avg Life Expectancy:  {df.loc[mask, "life_expectancy"].mean():.1f} years')
    print(f'  Avg Infant Mortality: {df.loc[mask, "infant_mortality"].mean():.1f} per 1,000')
    print()
````

**Interpretation (Part 3a):**

Fitting K-Means with `n_clusters=4` and `init='k-means++'` on the standardized 10-feature matrix assigns each of the ~160 countries to one of four clusters. The `k-means++` initialization selects starting centroids that are spread out in feature space, reducing the chance of poor local optima compared to random initialization. Setting `random_state=42` ensures the results are reproducible across runs. The four clusters should roughly stratify countries by development level - high-GDP, high-longevity economies in one group and high-mortality, low-GDP economies in another - though the algorithm finds these groupings purely from geometric distance in feature space, not from any predefined income labels.


````python
# ------------------------------
# ✏️ YOUR TASK - Fill in the blanks
# Step 3b: PCA projection to 2D and scatter plot
# ------------------------------

# Reduce 10 dimensions to 2 using PCA (n_components=2 for a 2D scatter plot)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# How much variance do the first 2 PCs explain?
print(f'Variance explained by PC1: {pca.explained_variance_ratio_[0]:.1%}')
print(f'Variance explained by PC2: {pca.explained_variance_ratio_[1]:.1%}')
print(f'Total variance explained:  {sum(pca.explained_variance_ratio_):.1%}')

# Scatter plot colored by cluster
fig, ax = plt.subplots(figsize=(10, 7))
colors = ['#e74c3c', '#3498db', '#2ecc71', '#f39c12']
cluster_names = ['Cluster 0', 'Cluster 1', 'Cluster 2', 'Cluster 3']

for k in range(4):
    mask = df['cluster'] == k
    ax.scatter(
        X_pca[mask, 0], X_pca[mask, 1],
        c=colors[k], label=cluster_names[k],
        alpha=0.7, edgecolors='white', s=60
    )

ax.set_xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.1%} variance)', fontsize=12)
ax.set_ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.1%} variance)', fontsize=12)
ax.set_title('K-Means Clustering of World Economies (K=4, PCA Projection)', fontsize=13)
ax.legend(fontsize=10)
plt.tight_layout()
plt.show()
````

**Interpretation (Part 3b - PCA Scatter Plot):**

PCA reduces the 10-dimensional standardized feature matrix to 2 dimensions by finding the orthogonal directions of maximum variance. PC1 typically captures the dominant development gradient - the axis along which GDP per capita, life expectancy, and internet access vary most together - while PC2 captures a secondary dimension such as trade openness or inequality. The scatter plot lets us visually assess whether K-Means produced well-separated clusters: if the four groups form distinct, non-overlapping regions in PCA space, the clusters are meaningful; if they heavily overlap, K=4 may be over- or under-fitting the natural structure. Note that PCA is used only for visualization here - K-Means itself clustered on all 10 standardized dimensions, not on the 2D projection.


## Part 4: Elbow Plot + Silhouette Analysis (YOUR TASK)

How do we know K=4 is the right number of clusters? Two diagnostic tools:

1. **Elbow method:** Plot WCSS (inertia) vs K. Look for where the rate of decrease slows.
2. **Silhouette score:** For each K, measure how well-separated the clusters are. Higher is better.

**Hints:**
- Loop over K = 2, 3, ..., 10
- WCSS is stored in `kmeans.inertia_` after fitting
- Silhouette: `silhouette_score(X_scaled, labels)`

````python
# ------------------------------
# ✏️ YOUR TASK - Fill in the blanks
# Step 4: Elbow plot and silhouette analysis
# ------------------------------

K_range = range(2, 11)
wcss_list = []
sil_list = []

for k in K_range:
    km = KMeans(
        n_clusters=k,
        init='k-means++',
        n_init='auto',
        random_state=42
    )
    labels = km.fit_predict(X_scaled)  # Cluster the standardized feature matrix

    # Append WCSS (stored in .inertia_ after fitting)
    wcss_list.append(km.inertia_)

    # Append silhouette score - measures cluster cohesion vs separation
    sil_list.append(silhouette_score(X_scaled, labels))

# Plot side by side
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Elbow plot
axes[0].plot(K_range, wcss_list, 'bo-', linewidth=2, markersize=8)
axes[0].set_xlabel('Number of Clusters (K)', fontsize=12)
axes[0].set_ylabel('Within-Cluster Sum of Squares (WCSS)', fontsize=12)
axes[0].set_title('Elbow Method', fontsize=13)

# Silhouette plot
axes[1].plot(K_range, sil_list, 'go-', linewidth=2, markersize=8)
axes[1].set_xlabel('Number of Clusters (K)', fontsize=12)
axes[1].set_ylabel('Mean Silhouette Score', fontsize=12)
axes[1].set_title('Silhouette Analysis', fontsize=13)

plt.tight_layout()
plt.show()

# Print scores
print('K  | WCSS       | Silhouette')
print('--|------|-----')
for k, w, s in zip(K_range, wcss_list, sil_list):
    print(f'{k}  | {w:>10.1f} | {s:.4f}')

best_k = list(K_range)[np.argmax(sil_list)]
print(f'\nBest K by silhouette score: {best_k}')
````

**Interpretation (Part 4 - Elbow & Silhouette):**

The elbow plot shows WCSS (inertia) falling steeply from K=2 to K=4 and flattening thereafter - the inflection around K=3 or K=4 signals that adding more clusters beyond that point yields diminishing reductions in within-cluster variance. The silhouette plot provides a complementary view: the silhouette score peaks near K=2 or K=3 because with fewer clusters, each cluster is internally tighter and more separated from others; as K increases, clusters begin splitting coherent groups and scores typically decline. Together, these diagnostics suggest K=3 or K=4 as reasonable choices for this dataset. K=4 aligns neatly with the World Bank's four income classifications, making it substantively defensible even if a slightly different K scores marginally higher on silhouette. The key takeaway is that there is no single 'correct' K - the elbow and silhouette are heuristics that inform judgment, not definitive answers.


## Part 5: Compare Clusters to World Bank Income Groups (YOUR TASK)

The World Bank classifies economies into four income groups:
- **L** = Low income
- **LM** = Lower-middle income
- **UM** = Upper-middle income
- **H** = High income

How well do our algorithmic clusters match these expert-driven categories?
Use a cross-tabulation (confusion matrix) to compare.

````python
# ------------------------------
# ✏️ YOUR TASK - Fill in the blanks
# Step 5: Cross-tabulate clusters vs World Bank income groups
# ------------------------------

# Fetch World Bank income group classification
income_groups = {}
for economy in wb.economy.list():
    income_groups[economy['id']] = economy.get('incomeLevel', {}).get('id', 'Unknown')

# Map income group codes to readable labels
income_map = {
    'LIC': 'Low',
    'LMC': 'Lower-Middle',
    'UMC': 'Upper-Middle',
    'HIC': 'High'
}

df['wb_income'] = df.index.map(income_groups)
df['wb_income'] = df['wb_income'].map(income_map).fillna('Unknown')

# Drop unknowns for comparison
df_compare = df[df['wb_income'] != 'Unknown'].copy()

# Cross-tabulation: algorithmic cluster (rows) vs World Bank income group (columns)
ct = pd.crosstab(
    df_compare['cluster'],    # K-Means algorithmic cluster assignment
    df_compare['wb_income'],  # World Bank official income group classification
    margins=True
)
print('=== Cross-Tabulation: K-Means Clusters vs World Bank Income Groups ===')
print(ct)
print()

# Visualize as a heatmap
fig, ax = plt.subplots(figsize=(8, 5))
ct_no_margins = ct.iloc[:-1, :-1]  # Remove margins for heatmap
sns.heatmap(ct_no_margins, annot=True, fmt='d', cmap='YlOrRd', ax=ax)
ax.set_xlabel('World Bank Income Group', fontsize=12)
ax.set_ylabel('K-Means Cluster', fontsize=12)
ax.set_title('Cluster vs Income Group Alignment', fontsize=13)
plt.tight_layout()
plt.show()

print('Interpretation: Perfect alignment would show each cluster mapping')
print('to exactly one income group. Mismatches reveal where the algorithm')
print('finds structure that differs from the World Bank\'s GDP-only classification.')
````

**Interpretation (Part 5 - Cluster vs Income Group):**

The cross-tabulation reveals the degree to which an unsupervised algorithm trained on 10 development indicators recovers the World Bank's expert income classifications. Where a single K-Means cluster maps predominantly to one income group (e.g., most High-income countries fall in Cluster X), the algorithm is effectively rediscovering that partition from raw data - a strong validation of both the features chosen and the K=4 specification. Mismatches are equally informative: countries that the World Bank classifies as Upper-Middle income but K-Means places in a High-income cluster may have strong health or connectivity indicators despite mid-range GDP, while countries clustered together across two income groups may reflect structural similarities that GDP thresholds alone obscure. This comparison underscores that K-Means clusters are data-driven constructs, not ground truth - they capture multidimensional proximity, whereas the World Bank's classification is anchored to a single GNI per capita threshold.

## Challenge: Cluster California Housing Census Tracts (Take-Home)

Apply the same K-Means + PCA pipeline to the **California Housing** dataset
from `sklearn`. Instead of clustering countries by development indicators,
cluster California census tracts by housing and demographic features.

**Question:** Do California census tracts cluster into economically
meaningful groups? Can you label each cluster (e.g., "urban high-income",
"rural low-income", "coastal expensive")?

````python
# ------------------------------
# ✏️ YOUR TASK - Challenge: California Housing Clustering
# Apply the full pipeline: load → standardize → K-Means → PCA → interpret
# ------------------------------

from sklearn.datasets import fetch_california_housing

# Step A: Load data
cal = fetch_california_housing(as_frame=True)
cal_df = cal.frame
print(f'Shape: {cal_df.shape}')
print(f'Features: {list(cal_df.columns)}')
print()

# Step B: Select features for clustering (exclude target)
cal_features = cal_df.drop(columns=['MedHouseVal'])

# Step C: Standardize - required before K-Means to equalize feature scales
cal_scaler = StandardScaler()
cal_scaled = cal_scaler.fit_transform(cal_features)
print(f'Standardized feature matrix shape: {cal_scaled.shape}')
print(f'Feature means (should be ~0): {cal_scaled.mean(axis=0).round(3)}')
print()

# Step D: Find optimal K via elbow + silhouette for K=2..8
cal_K_range = range(2, 9)
cal_wcss = []
cal_sil = []

for k in cal_K_range:
    km_cal = KMeans(n_clusters=k, init='k-means++', n_init='auto', random_state=42)
    lbl = km_cal.fit_predict(cal_scaled)
    cal_wcss.append(km_cal.inertia_)
    cal_sil.append(silhouette_score(cal_scaled, lbl))

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
axes[0].plot(cal_K_range, cal_wcss, 'bo-', linewidth=2, markersize=8)
axes[0].set_xlabel('Number of Clusters (K)', fontsize=12)
axes[0].set_ylabel('WCSS (Inertia)', fontsize=12)
axes[0].set_title('Elbow Method - California Housing', fontsize=13)

axes[1].plot(cal_K_range, cal_sil, 'go-', linewidth=2, markersize=8)
axes[1].set_xlabel('Number of Clusters (K)', fontsize=12)
axes[1].set_ylabel('Mean Silhouette Score', fontsize=12)
axes[1].set_title('Silhouette Analysis - California Housing', fontsize=13)
plt.tight_layout()
plt.show()

cal_best_k = list(cal_K_range)[np.argmax(cal_sil)]
print(f'Best K by silhouette: {cal_best_k}')
print(f'Silhouette scores: {[round(s, 4) for s in cal_sil]}')
print()

# Step E: Fit K-Means with chosen K (using 4 for interpretability)
chosen_k = 4  # Balances silhouette performance with interpretability
km_final = KMeans(n_clusters=chosen_k, init='k-means++', n_init='auto', random_state=42)
cal_df['cluster'] = km_final.fit_predict(cal_scaled)
print(f'Cluster sizes (K={chosen_k}):')
print(cal_df['cluster'].value_counts().sort_index())
print()

# Step F: PCA 2D visualization
cal_pca = PCA(n_components=2)
cal_X_pca = cal_pca.fit_transform(cal_scaled)

print(f'Variance explained by PC1: {cal_pca.explained_variance_ratio_[0]:.1%}')
print(f'Variance explained by PC2: {cal_pca.explained_variance_ratio_[1]:.1%}')
print(f'Total variance explained:  {sum(cal_pca.explained_variance_ratio_):.1%}')
print()

fig, ax = plt.subplots(figsize=(10, 7))
cal_colors = ['#e74c3c', '#3498db', '#2ecc71', '#f39c12']
for k in range(chosen_k):
    mask = cal_df['cluster'] == k
    ax.scatter(
        cal_X_pca[mask, 0], cal_X_pca[mask, 1],
        c=cal_colors[k], alpha=0.4, s=10, label=f'Cluster {k}'
    )
ax.set_xlabel(f'PC1 ({cal_pca.explained_variance_ratio_[0]:.1%} variance)', fontsize=12)
ax.set_ylabel(f'PC2 ({cal_pca.explained_variance_ratio_[1]:.1%} variance)', fontsize=12)
ax.set_title(f'K-Means Clustering of California Census Tracts (K={chosen_k}, PCA Projection)', fontsize=13)
ax.legend(fontsize=10)
plt.tight_layout()
plt.show()

# Step G: Cluster summary table with economically meaningful labels
cluster_labels = {
    0: 'TBD after viewing centroids',
    1: 'TBD after viewing centroids',
    2: 'TBD after viewing centroids',
    3: 'TBD after viewing centroids',
}

summary_cols = ['MedInc', 'HouseAge', 'AveRooms', 'AveOccup', 'Latitude', 'Longitude']
# Add MedHouseVal back for interpretation (it was excluded from clustering)
cal_df['MedHouseVal'] = cal.frame['MedHouseVal']

print('=== Cluster Summary ===')
cluster_summary = cal_df.groupby('cluster')[summary_cols + ['MedHouseVal']].mean().round(2)
print(cluster_summary.to_string())
print()

# Assign economically meaningful labels based on observed centroids
# High MedInc + high MedHouseVal + coastal latitude → 'Affluent Coastal'
# Low MedInc + high HouseAge + dense occupancy  → 'Urban Working Class'
# Low MedInc + rural latitude (inland)           → 'Rural Low-Income'
# Mid MedInc + newer homes                       → 'Suburban Middle Class'

# Re-label based on median income rank across clusters
income_rank = cluster_summary['MedInc'].rank()
label_map = {}
for k in range(chosen_k):
    rank = income_rank[k]
    lat = cluster_summary.loc[k, 'Latitude']
    age = cluster_summary.loc[k, 'HouseAge']
    if rank == chosen_k:  # Highest income
        label_map[k] = 'Affluent Coastal'
    elif rank == 1:       # Lowest income
        label_map[k] = 'Rural / Low-Income Inland'
    elif age > 25:
        label_map[k] = 'Urban Working Class (Older Stock)'
    else:
        label_map[k] = 'Suburban Middle Class'

print('=== Economically Meaningful Cluster Labels ===')
for k, label in label_map.items():
    n = (cal_df['cluster'] == k).sum()
    med_inc = cluster_summary.loc[k, 'MedInc']
    med_val = cluster_summary.loc[k, 'MedHouseVal']
    print(f'Cluster {k} ({n:,} tracts): "{label}"')
    print(f'  Avg Median Income: ${med_inc:.2f} (x$10k) | Avg Median House Value: ${med_val:.3f} (x$100k)')
````

**Interpretation (Challenge - California Housing Clustering):**

Applying the same standardize → K-Means → PCA pipeline to 20,640 California census tracts produces economically interpretable groupings. The elbow and silhouette diagnostics suggest K=2 or K=3 as statistically optimal - the dataset has a strong first principal component capturing the income/value gradient - but K=4 is used here for richer granularity and comparability with the World Bank analysis. The PCA scatter plot reveals one densely packed cluster (likely lower-income inland tracts) and a more dispersed high-income cluster, consistent with California's sharp geographic wealth divide. Cluster labels are assigned post-hoc by inspecting centroid values: tracts with high median income and coastal latitude map to 'Affluent Coastal', low-income inland tracts form a 'Rural / Low-Income Inland' group, older dense housing stock clusters as 'Urban Working Class', and newer mid-income suburbs form a 'Suburban Middle Class' group. These labels are interpretive - the algorithm knows nothing about geography or income categories; it only minimizes Euclidean distance in standardized feature space - but the fact that meaningful socioeconomic patterns emerge validates that the eight housing features contain real signal about neighborhood type.
