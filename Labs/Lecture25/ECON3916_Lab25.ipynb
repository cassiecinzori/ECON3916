## Part 4: Add a Custom Section (YOUR TASK + AI CO-PILOT - 10 min)

**Chosen option: Option A - Data Projects Gallery**

### Prompts used:

**Prompt 1 (initial):**
```
Add a Data Projects Gallery section below the Projects section. It should show
screenshot thumbnails from my lab outputs in a responsive grid. Each card should
have a hover effect that reveals the project title and a short description.
Add filter buttons at the top to filter cards by technology: Python, scikit-learn,
NLP, Time Series, Causal ML. Use my existing color palette.
```

**What Claude Code generated on the first attempt:**
A `ProjectGallery.tsx` component with a CSS grid, hover overlay using absolute positioning,
and a filter bar. The structure was correct but it referenced a non-existent `useFilter` hook
which caused an immediate import error. Image aspect ratios were also inconsistent.

**What I had to revise:**
- Replaced the hallucinated `useFilter` hook with a simple `useState` for the active filter
- Added `object-fit: cover` with a fixed container height to normalize image display
- Adjusted hover transition timing (the default 500ms felt sluggish; changed to 200ms)

**Prompt 2 (fix):**
```
The useFilter import is failing - there is no such hook. Replace the filter logic
with a simple useState for activeFilter (default: 'All'). Also add object-fit: cover
and a fixed height of 200px to the image containers so thumbnails are uniform.
```

**Iterations required:** 2 prompts to get a working result, 1 more minor tweak for transition speed.

### Q1: What worked well with AI-assisted development?

*Think about: speed of scaffolding, code quality, design suggestions, handling of repetitive tasks.*

**Your answer:**
The scaffolding step was genuinely fast in a way that would have taken hours manually. Within a few minutes of providing my background and style preferences, Claude Code had generated a complete Next.js project with component files, Tailwind configuration, and a working dev server - the kind of boilerplate that is tedious to write but straightforward to evaluate once it exists. It also handled the repetitive parts of the project cards well: once I gave it a template for one project entry (title, description, tech tags, GitHub link), it applied the pattern consistently across all three without me having to specify the structure each time. The dark/light mode toggle and responsive layout were both implemented correctly on the first attempt, which would have required me to look up Tailwind's `dark:` variant syntax and test breakpoints myself.

### Q2: What did you have to fix manually?

*Think about: layout bugs, incorrect assumptions, styling mismatches, missing features, hallucinated APIs.*

**Your answer:**
The Data Projects Gallery required the most iteration. Claude Code's first attempt used a CSS grid with hover overlays, but the thumbnails were rendering at inconsistent aspect ratios because it assumed all images would be the same size, which mine were not. I had to prompt it specifically to use `object-fit: cover` with a fixed container height to normalize them, which it then applied correctly. It also initially generated a filter component that referenced a `useFilter` hook that does not exist in any library - I caught this because the dev server threw an import error immediately, but it is a good reminder that plausible-looking code is not the same as working code. The color palette it chose on the first pass was generic; I had to give it explicit hex values from my style reference to get the exact look I wanted rather than accepting its default navy-and-white scheme.

### Q3: How does AI-assisted development compare to writing code entirely yourself?

*Think about: the HADI loop from Chapter 25 (Hypothesis, Action, Data, Insight). Did you follow this pattern? Where did you need domain knowledge that the AI lacked?*

**Your answer:**
The HADI loop described in Chapter 25 maps well onto how this session actually ran: I had a hypothesis about what the site should look like, the action was prompting Claude Code, the data was the rendered output at localhost:3000, and the insight informed my next prompt revision. What changed compared to writing code myself is where the cognitive effort goes - instead of thinking about syntax and component structure, I was thinking about what to ask for and whether the output matched my intent, which is a higher-level skill but still a skill. The places where my domain knowledge mattered most were in the project descriptions themselves (Claude Code had no way to know what my actual R-squared values were or which methods I used) and in evaluating whether the generated code had bugs, since accepting output uncritically would have left the broken `useFilter` hook in the production build. Writing code entirely myself would have been slower but would have left me with a deeper understanding of the codebase; the tradeoff is real and context-dependent.

### Q4: Prompt engineering lessons

*What makes a good prompt for Claude Code? What makes a bad one? Give one specific example of a prompt you improved.*

**Your answer:**
Good prompts for Claude Code are specific about constraints, outputs, and context simultaneously - they tell the AI what file to edit, what the end result should look like, and what existing structure to preserve. Bad prompts are vague about scope: early on I asked it to "make the projects section look better," which produced a complete restyling I did not want because it had no signal about what "better" meant to me. I improved that prompt to: "In the ProjectCard component, increase the padding inside the card to 24px, change the tag background to `bg-slate-100 dark:bg-slate-700`, and keep everything else the same" - that level of specificity produced exactly the change I wanted with no unintended side effects. The general pattern I learned is to decompose broad aesthetic goals into concrete CSS or layout changes before prompting, rather than hoping the AI will interpret a subjective preference correctly.

### Q5: If you were hiring a junior data scientist, would their portfolio site influence your evaluation? Why or why not?

*Connect to the "signaling" concept from labor economics -- what does a portfolio signal that a resume alone does not?*

**Your answer:**
Yes, a portfolio site would meaningfully influence my evaluation because it signals things a resume structurally cannot. A resume lists tools and project titles, but a portfolio with working GitHub links and visible outputs demonstrates that the candidate can take an analysis from a Jupyter notebook to something presentable to a non-technical audience - which is most of what junior analysts actually spend their time doing. In the signaling framework from labor economics, the portfolio is a costly signal in the Spence sense: building and deploying a site requires real effort, so its existence screens for candidates who care about communication quality and follow through on non-required work. For a quantitative role specifically, I would look at whether the project descriptions show genuine interpretation of results rather than just listing methods used, since the ability to connect a model output to an economic conclusion is the skill that is hardest to teach and easiest to distinguish from someone who can run code but cannot explain what it means.
