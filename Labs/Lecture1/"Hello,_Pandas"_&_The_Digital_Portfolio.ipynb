### Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Manual Dictionary Construction

```python
data = {
    'Country': ['United States','Argentina', 'Australia', 'Brazil', 'Britain',
                'China', 'Egypt', 'Euro area', 'Hong Kong', 'Indonesia',
                'Japan', 'Mexico', 'Norway', 'Pakistan', 'Philippines',
                'Russia', 'Saudi Arabia', 'South Africa', 'South Korea'],

    'iso_a3': ['USA', 'ARG', 'AUS', 'BRA', 'GBR', 'CHN', 'EGY', 'EUZ', 'HKG',
               'IDN', 'JPN', 'MEX', 'NOR', 'PAK', 'PHL', 'RUS', 'SAU', 'ZAF',
               'KOR'],

   'local_price': [4.79, 28, 5.3, 13.5, 2.89, 17.2, 16.93,
                3.68, 18.8, 27939, 370, 49, 48, 300, 163,
                89, 11, 25.5, 4100],

   'dollar_ex': [1.0, 8.62, 1.23, 2.59, 0.66, 6.21, 7.36, 0.86,
              7.74, 12472, 117.83, 14.63, 7.62, 100.67,
              44.41, 65.44, 3.75, 11.49, 1084.66]
  }


# Convert to df
df = pd.DataFrame(data)
print(df)
```

###  Calculate Implied PPP and Valuation

```python
# 1. Calculate the price in USD
df['dollar_price'] = df['local_price'] / df['dollar_ex']

# 2. Identify the US Price (benchmark)
us_price = df.loc[df['iso_a3'] == 'USA', 'dollar_price'].values

# 3. Calculate Implied PPP
df['implied_ppp'] = df['local_price'] / us_price

# 4. Calculate Valuation
df['valuation'] = (df['implied_ppp'] - df['dollar_ex']) / df['dollar_ex']

# 5. Convert to Percentage
df['valuation_pct'] = df['valuation'] * 100

# Print the updated DataFrame
print(df[['Country', 'dollar_price', 'valuation_pct']].sort_values('valuation_pct'))

```

### Visualization

```python
plt.figure(figsize=(10,6))

# Create bar plot
sns.barplot(x='valuation_pct',
            y='Country',
            data=df.sort_values('valuation_pct')
            palette='vlag')

plt.axvline(x=0, color='black', linestyle='--')

plt.title('The Big Mac Index: Currency Valuation vs USD (Jan 2025)')
plt.xlabel('% Under/Over Value Relative to USDA')
plt.ylabel('Country')
plt.show()
```
